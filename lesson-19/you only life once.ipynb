```python
from ultralytics import YOLO
import os
from dotenv import load_dotenv
from datetime import datetime as dt
import warnings

load_dotenv("../.env")
```

```python
vid1 = os.getenv("vid1")
vid2 = os.getenv("vid2")
vid3 = os.getenv("vid3")
vid4 = os.getenv("vid4")
_str = dt.now()
```

```python
model = YOLO('yolov8m.pt')
```

```python
class_list = []
for _ in range(len(model.names)):
    if _ != 38:
        class_list.append(_)
    else:
        continue
```

```python
print(f"starting at {_str}")
results = model.predict(
    source=vid1, 
    show=True,
    conf=0.3,
    stream=True,
    imgsz=480,
    classes=class_list,
    device='cuda:0'
)

print(f"end, time: {dt.now() - _str}")
```

```python
for r in results:
    boxes = r.boxes.xywh.cpu().numpy()
    num_persons = (r.boxes.cls.cpu().numpy() == 0).sum()
```

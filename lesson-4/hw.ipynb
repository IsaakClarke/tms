```python
import numpy as np

A4 = np.array([[1, -2, 8],
               [7, 5, -2],
               [-2, 3, 0]])

B4 = np.array([[0, 2, 4],
               [1, -3, -2],
               [-2, 1, 3]])

C4 = np.array([[0, 3],
               [5, -4],
               [7, -1]])

print("=== Вариант 4 ===")
print("A + B:\n", A4 + B4)
print("\nB - A:\n", B4 - A4)
print("\nA @ C:\n", A4 @ C4)
print("\nA @ B @ C:\n", A4 @ B4 @ C4)
```

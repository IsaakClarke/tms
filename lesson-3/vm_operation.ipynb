```python
class vector():
    def __init__(self, data, multi = 1):
        if not isinstance(data, list):
            raise ValueError("!= list[]")
        if not all(isinstance(x, (int, float)) for x in data):
            raise ValueError("data value != (float, int)")
        if not isinstance(multi, (float, int)):
            raise ValueError("multi != (float, int)")
        self.data = data
        self.multi = multi
    
    def __len__(self):
        return len(self.data)

    def get_sum(self):
        return sum(self.data)

    def get_min(self):
        return min(self.data)

    def get_max(self):
        return max(self.data)

    def get_l2(self):
        total = sum(x**2 for x in self.data)
        return round(total**0.5, 2)

    def get_normal(self):
        l2 = self.get_l2()
        return self.data[:] if l2 == 0 else [round(x/l2, 2) for x in self.data]

    def get_scale(self):
        return [x * self.multi for x in self.data]
    
    def __repr__(self):
        return f"Vector({self.data})"
            
class matrix(vector):
    def __init__(self, data, multi = 1):
        if not (isinstance(data, list) and all(isinstance(row, list) for row in data)):
            raise ValueError("data != list[list[]]")
        if not all(all(isinstance(x, (float, int)) for x in row) for row in data):
            raise ValueError("data value != (float, int)")
        if not (isinstance(multi, (int, float))):
            raise ValueError("mult != (float, int)")
            
        row_len = len(data[0])
        
        if not all(len(row) == row_len for row in data):
            raise ValueError("len(list[0]) != len(list[1])")
        self.data = data
        self.multi = multi
        

    def get_shape(self):
        return str(len(self.data)) + 'x' + str(len(self.data[0]))

    def get_trans(self):
        t = []
        for j in range(len(self.data[0])):
            new_row = []
            for i in range(len(self.data)):
                new_row.append(self.data[i][j])
            t.append(new_row)
        return t

    def get_trace(self):
        if len(self.data) != len(self.data[0]):
            raise ValueError("Матрица не квадрат")
        return sum(self.data[i][i] for i in range(len(self.data)))

    def get_det(self):
        if len(self.data) != len(self.data[0]):
            raise ValueError("Матрица не квадрат")

        n = len(self.data)

        if n == 1:
            return self.data[0][0]
        elif n == 2:
            return self.data[0][0] * self.data[1][1] - self.data[0][1]*self.data[1][0]
        elif n > 2:
            raise NotImplementedError("Не смог в детерминант матриц len() > 2")

    def get_scale(self):
        return [[x*self.multi for x in row] for row in self.data]
    
    def __str__(self):
        return '\n'.join(' '.join(str(x) for x in row) for row in self.data)

    def __repr__(self):
        return f"Matrix({self.data})"

m = matrix([[1, 4], [2, 3]])
v = vector([1, 2])

print(f"Длина: {len(v)}, сумма: {v.get_sum()}, min: {v.get_min()}, max: {v.get_max()}, евклидова норма: {v.get_l2()}, нормализация: {v.get_normal()}, скал. произв: {v.get_scale()}")
print(f"Размер: {m.get_shape()}, Транспон.: {m.get_trans()}, След: {m.get_trace()}, Опр.: {m.get_det()}, Скал. произв.: {m.get_scale()}")
```

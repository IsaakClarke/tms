```python
xl = x.values.tolist()
yl = y.values.tolist()
```

```python
idx = list(range(len(xl)))
```

```python
import random as rnd

random.shuffle(idx)

xl_shuffle = [xl[i] for i in idx]
yl_shuffle = [yl[i] for i in idx]
```

```python
split_idx = int(0.8 * len(xl_shuffle))

xl_train = xl_shuffle[:split_idx]
yl_train = yl_shuffle[:split_idx]

xl_test = xl_shuffle[split_idx:]
yl_test = yl_shuffle[split_idx:]
```

```python
w = [rnd.random() for i in range(len(xl[0]))]
```

```python
b = 0
```

```python
lr = 0.000001
epoch = 10000
L2 = 0.000001
```

```python
import math

def pred(xl, w, b):
    res = sum(w[i] * xl[i] for i in range(len(xl)))
    return 1/(1 + math.exp(-(res + b)))
```

```python
def learn(xl, w, b, yl, epoch, lr, L2):
    for i in range(epoch):
        
        y_p = []
        errors = []
        
        for j in range(len(xl)):
            a = pred(xl[j], w, b)
            y_p.append(a)
            errors.append(a - yl[j])
            for k in range(len(xl[j])):
                w[k] -= lr * ((a - yl[j]) * xl[j][k] + L2 * w[k])
            
            b -= lr * (a - yl[j])
            
    return w, b
```

```python
learn(xl_train, w, b, yl_train, epoch, lr, L2)
```

```python
# MSE train
predict = []
error = []
error_sq = []

predict = [pred(i, w, b) for i in xl_train]

for i in range(len(predict)):
    error.append(predict[i] - yl_train[i])

for i in range(len(error)):
    error_sq.append(error[i]**2)

MSE = sum(error_sq)/len(error_sq)

print(MSE)
```

```python
# MSE test
predict = []
error = []
error_sq = []

predict = [pred(i, w, b) for i in xl_test]

for i in range(len(predict)):
    error.append(predict[i] - yl_test[i])

for i in range(len(error)):
    error_sq.append(error[i]**2)

MSE = sum(error_sq)/len(error_sq)

print(MSE)
```

```python
pred_train = []
for row in xl_train:
    res = sum(w[i]*row[i] for i in range(len(row)))
    pred_train.append(1/(1 + math.exp(-(res + b))))

pred_test = []
for row in xl_test:
    res = sum(w[i]*row[i] for i in range(len(row)))
    pred_test.append(1/(1 + math.exp(-(res + b))))
```

```python
threshold = 0.5
```

```python
pred_train_p = [1 if p >= 0.5 else 0 for p in pred_train]
```

```python
pred_test_p = [1 if p >= 0.5 else 0 for p in pred_test]
```

```python
corr = sum([1 for i in range(len(pred_train_p)) if pred_train_p[i] == yl_train[i]])
accuracy_train = corr / len(yl_train)
print(f"train {accuracy_train}")

corr = sum([1 for i in range(len(pred_test_p)) if pred_test_p[i] == yl_test[i]])
accuracy_test = corr / len(yl_test)
print(f"test {accuracy_test}")
```

```python
def tp_fp_fn(y_true, y_pred):
    
    TP = sum([1 for i in range(len(y_true)) if y_true[i] == 1 and y_pred[i] == 1])
    FP = sum([1 for i in range(len(y_true)) if y_true[i] == 0 and y_pred[i] == 1])
    FN = sum([1 for i in range(len(y_true)) if y_true[i] == 1 and y_pred[i] == 0])
    
    return TP, FP, FN

TP, FP, FN = tp_fp_fn(yl_train, pred_train_p)
precision_train = TP / (TP + FP) if (TP + FP) > 0 else 0
recall_train = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_train = 2 * precision_train * recall_train / (precision_train + recall_train) if (precision_train + recall_train) > 0 else 0

print(f"train precision {precision_train}")
print(f"train recall {recall_train}")
print(F"train F1 {f1_train}")
```

```python
TP, FP, FN = tp_fp_fn(yl_test, pred_test_p)
precision_test = TP / (TP + FP) if (TP + FP) > 0 else 0
recall_test = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_test = 2 * precision_test * recall_test / (precision_test + recall_test) if (precision_test + recall_test) > 0 else 0

print(f"train precision {precision_test}")
print(f"train recall {recall_test}")
print(F"train F1 {f1_test}")
```
